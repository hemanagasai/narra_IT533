```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import statistics
```

```python
hm =pd.read_csv("https://raw.githubusercontent.com/sekharmm/mandava_IT533/main/amazon_toy_sample_clean.csv",encoding="ISO-8859-1")
```

```python
# Calculate min and max values
df_min, df_max = np.min(hm.average_review_rating_of_5),np.max(hm.average_review_rating_of_5)
## Calculate Quartiles
quartiles = np.percentile(hm.average_review_rating_of_5, [25, 50, 75])
## Standard Deviation
s_dev = statistics.stdev(hm.average_review_rating_of_5)
## Summary
print("Min Value: ",df_min)
print("Max Value: ",df_max)
print("Lower Quartile: ",quartiles[0])
print("Median: ",quartiles[1])
print("Upper Quartile: ",quartiles[2])
print("Standard Deviation: ", s_dev)
```
